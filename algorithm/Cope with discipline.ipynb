```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./data/repa_data_20230630.xlsx')
df.head()
```

```python
dis_set = np.unique(df['discipline'].to_numpy())
```

```python
dis_dict = {}
for dis in dis_set:
    dis_arr = df[df['discipline'] == dis]
    total_len = dis_arr.shape[0]
    re_len = dis_arr[dis_arr['situation'] == 1].shape[0]
    dis_dict[dis] = [re_len / total_len]
```

```python
dis_dict['0']
```

```python
dis_df = pd.DataFrame(dis_dict)
dis_df.head()
```

```python
dis_df.to_excel('./data/discipline.xlsx')
```
