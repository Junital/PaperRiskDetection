```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("../data/repa_data_20231022.csv")
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.matplotlib_fname()
```

```python
years = df["published_time"].tolist()
```

```python
filter_years = [x for x in years if x != 0]
```

```python
sns.set_theme()
sns.histplot(filter_years)
plt.rcParams['font.sans-serif'] = ['STSong']
plt.xlabel("论文发表年份")
plt.ylabel("论文数量")
plt.title("论文发表年份直方图")
```

```python
is_Re = df["is_retracted"].tolist()
```

```python
filter_situa = []
for i in range(0, len(years)):
    if(years[i] != 0):
        filter_situa.append(is_Re[i])
```

```python
sns.boxplot(x=filter_years, y=filter_situa)
plt.xlabel("论文发表年份")
plt.ylabel("是否为撤销论文")
plt.title("不同种类论文发表年份的箱线图")
```
