```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

df = pd.read_csv("../data/repa_data_20231022.csv")

years = df["published_time"].tolist()
filter_years = [x for x in years if x != 0]

is_Re = df["is_retracted"].tolist()
filter_situa = []
for i in range(0, len(years)):
    if(years[i] != 0):
        filter_situa.append(is_Re[i])
```

```python
tot_com = df["total_comment"].tolist()
ari = df["author_active"].tolist()
com_len = df["comment_len"].tolist()
```

```python
filter_totcom = []
filter_ari = []
filter_comlen = []
for i in range(0, len(years)):
    if(years[i] != 0):
        filter_totcom.append(tot_com[i])
        filter_ari.append(ari[i])
        filter_comlen.append(com_len[i])
```

```python
sns.set_theme()
plt.rcParams['font.sans-serif'] = ['STSong']
fig, axes = plt.subplots(3, 1, figsize=(8, 6))

sns.boxplot(x=filter_comlen, y=filter_situa, ax=axes[0])
sns.boxplot(x=filter_ari, y=filter_situa, ax=axes[1])
sns.boxplot(x=filter_totcom, y=filter_situa, ax=axes[2])

axes[0].set_title("评论长度")
axes[1].set_title("作者回复积极性")
axes[2].set_title("评论回复总数")

plt.tight_layout()  # 调整子图的布局
plt.show()
```
