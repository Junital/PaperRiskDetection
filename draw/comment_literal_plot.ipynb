```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

df = pd.read_csv("../data/repa_data_20231022.csv")

years = df["published_time"].tolist()
filter_years = [x for x in years if x != 0]

is_Re = df["is_retracted"].tolist()
filter_situa = []
for i in range(0, len(years)):
    if(years[i] != 0):
        filter_situa.append(is_Re[i])
```

```python
reword = df["reword"].tolist()
com_pos = df["my_prediction_POS"].tolist()
com_neu = df["my_prediction_NEU"].tolist()
com_neg = df["my_prediction_NEG"].tolist()
```

```python
filter_reword = []
filter_pos = []
filter_neu = []
filter_neg = []
for i in range(0, len(years)):
    if(years[i] != 0):
        filter_reword.append(reword[i])
        filter_pos.append(com_pos[i])
        filter_neu.append(com_neu[i])
        filter_neg.append(com_neg[i])
```

```python
sns.set_theme()
plt.rcParams['font.sans-serif'] = ['STSong']
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

sns.boxplot(x=filter_reword, y=filter_situa, ax=axes[0, 0])
sns.boxplot(x=filter_pos, y=filter_situa, ax=axes[0, 1])
sns.boxplot(x=filter_neu, y=filter_situa, ax=axes[1, 0])
sns.boxplot(x=filter_neg, y=filter_situa, ax=axes[1, 1])

axes[0, 0].set_title("评论撤销词频分数")
axes[0, 1].set_title("积极情感分数")
axes[1, 0].set_title("中性情感分数")
axes[1, 1].set_title("消极情感分数")

plt.tight_layout()  # 调整子图的布局
plt.show()
```
